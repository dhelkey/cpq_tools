```python
#Imports
import networkx as nx
import matplotlib.pyplot as plt

#CPQCC Packages
from cpq_tools import compute_network
from cpq_tools.data import SAMPLE_NETWORK_DATA

#Display available sample network types
[key for key in SAMPLE_NETWORK_DATA.keys()]
```

```python
#Paramaters
#Manually specify positions (by node key)
pos = {
    '1': (0.1, 0.1), '2': (1, 0.3), '3': (1.75, 0.25),
    '4': (0, 1), '5': (.9, 1.25), '6': (2, 1),
    '7': (0.25, 1.75), '8': (1, 2), '9': (2, 2)
}

##FOr use with this data
# network_type = 'more_dense' 
# network_data = SAMPLE_NETWORK_DATA[network_type]
```

```python

for network_type, network_data in SAMPLE_NETWORK_DATA.items():

    network = compute_network(network_data, 'from_hosp','to_hosp')
    #Print data
    #network['metrics']
    #network_data.head(2)

    #Generateing images for each graph type


    # Drawing the graph with fixed positions
    nx.draw(network['graph_networkx'], pos,
            with_labels=True, node_color='lightblue', 
            node_size=800, font_size=10, font_weight='bold')
    plt.title(network_type)
    plt.show()
```
